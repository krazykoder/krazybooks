```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Define image path
image_path = './coins_image.png'
# image_path = './watershed_coins_01.jpg'
```

```python
cap = cv2.imread(image_path)
plt.imshow(cap)
## ignore explicits if not needed
# cap.set(cv2.cv.CV_CAP_PROP_FRAME_WIDTH, 1280)
# cap.set(cv2.cv.CV_CAP_PROP_FRAME_HEIGHT, 720)

# while(True):
# ret, frame = cap.read()
# roi = frame[0:500, 0:500]
roi = cap.copy()
gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
# gray_blur = cv2.blur(gray, (15, 15), 0)
gray_blur = cv2.GaussianBlur(gray, (15, 15), 0)
# gray_blur = cv2.bilateralFilter(gray, 9, 75,75)
# gray_blur = cv2.medianBlur(gray, 15)


thresh = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                               cv2.THRESH_BINARY_INV, 11, 1)

kernel = np.ones((1, 1), np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=3)

cont_img = closing.copy()
contours, hierarchy = cv2.findContours(cont_img, cv2.RETR_EXTERNAL,
                                       cv2.CHAIN_APPROX_SIMPLE)
```

```python
edged = cv2.Canny(gray, 75, 200)
plt.imshow(edged, cmap='gray')
```

```python
plt.imshow(thresh, cmap='gray')
```

```python
plt.imshow(gray_blur, cmap='gray')
```

```python
plt.imshow(closing, cmap='gray')
```

```python
n, h = np.histogram(gray, bins='auto')
plt.plot(n)
```

```python
n, h = np.histogram(gray_blur, bins='auto')
plt.plot(n)
```

```python
for cnt in contours:
    area = cv2.contourArea(cnt)
    if area < 2000 : #or area > 4000:
        continue

    if len(cnt) < 5:
        continue

    ellipse = cv2.fitEllipse(cnt)
    cv2.ellipse(roi, ellipse, (0,255,0), 2)
```

```python
cv2.imwrite("Morphological-Closing.png", closing)
cv2.imwrite("Adaptive-Thresholding.png", thresh)
cv2.imwrite('Contours.png', roi)
```
