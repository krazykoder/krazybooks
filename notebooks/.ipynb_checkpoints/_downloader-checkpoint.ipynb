```python
from  PIL import Image, ImageEnhance
import requests
from io import BytesIO
```

```python
url = "http://blog.christianperone.com/wp-content/uploads/2014/06/gray_image.png"
```

```python
response = requests.get(url)

img = Image.open(BytesIO(response.content))
img
```
