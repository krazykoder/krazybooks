## Clean up the links with sub references '#' from within document of the same page 

```python
for artist_name in artist_name_list_items:
    names = artist_name.get('href') #.contents[0]
    if '#' not in names: 
        print('http://interactivepython.org/runestone/static/pythonds/'+ names)
```

## Prepend root url if link does not start with 'http' or 'https' 

```python
for artist_name in artist_name_list_items:
    names = artist_name.get('href') #.contents[0]  
    if '#' not in names: 
        print(urlparse.urljoin(url, names)) # joins at last '/' in the base (1st arg).
```

```python
# Find root path 
root = url
if not url.endswith("/", 0) : 
    root = url.rpartition('/') 
print (root)
print ("Root url = " + root[0])
# root = root.geturl()

for artist_name in artist_name_list_items:
    names = artist_name.get('href') #.contents[0]  
#     if '#' not in names: 
    print(urlparse.urljoin(root[0]+'/', names, allow_fragments=1))
```
